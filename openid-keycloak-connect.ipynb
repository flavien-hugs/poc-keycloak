```python
from keycloak import KeycloakOpenID
```

```python
# INIT INSTANCE KEYCLOAK OPENID

kc_openid = KeycloakOpenID(
    server_url="http://0.0.0.0:8088",
    realm_name="auth",
    client_id="account",
    client_secret_key="bWUrxqhCIrBFwZAMTAllEJRg0bZebGRT",
    timeout=300,
)
```

```python
# VERIFICATION DE LA CONNEXION AU SERVEUR KEYCLOAK

kc_openid_instance = await kc_openid.a_well_known()
kc_openid_instance
```

```python
# LOGIN USER

result = await kc_openid.a_token(username="auth", password="admin")
result
```

```python
# LOGOUT USER

refresh_token = result["refresh_token"]
result = await kc_openid.a_logout(refresh_token=refresh_token)
result
```

```python
# REFRESH TOKEN USER

refresh_token = result["refresh_token"]
result = await kc_openid.a_refresh_token(refresh_token=refresh_token)
result
```

```python
# INTROSPECT TOKEN

access_token = result["access_token"]
result = await kc_openid.a_introspect(token=access_token)
result
```
